```python
from nba_api.stats.endpoints import playercareerstats, playergamelog, playergamelogs
from nba_api.stats.static import players


def get_first_and_last_season(id):
    career = playercareerstats.PlayerCareerStats(player_id=id) 
    player = players.find_player_by_id(player_id=id)
    player_name = player['full_name']
    # print(player)
    
    s = career.get_data_frames()[0]
    
    num_row, num_col = s.shape
    
    # print(num_row, num_col)
    
    first_season = s.loc[0, "SEASON_ID"]
    last_season = s.loc[num_row - 1, "SEASON_ID"]
    
    # print("first season: " + first_season)
    # print("last season: " + last_season)

    return (player_name, first_season, last_season)

get_first_and_last_season("203999")
```

```python
from nba_api.stats.endpoints import playergamelog, playergamelogs

def get_game_log_of_player(id):
    name, fs, ls = get_first_and_last_season(id)

    df = None

    start_year = int(fs.split('-')[0])
    end_year = int(ls.split('-')[0])

    for i in range(start_year, end_year + 1):
        year_query = str(i) + "-" + str((i + 1) % 100)
        print(year_query)

        stats = playergamelogs.PlayerGameLogs(player_id_nullable=id, season_nullable=year_query, season_type_nullable="Regular Season")
        pstats = playergamelogs.PlayerGameLogs(player_id_nullable=id, season_nullable=year_query, season_type_nullable="Playoffs")


        s = stats.get_data_frames()[0]
        ps = pstats.get_data_frames()[0]
        
        if not s.empty:
            if df is None:
                df = s
            else:
                df = pd.concat([df, s], axis=0, ignore_index=True)
        if not ps.empty:
            if df is None:
                df = ps
            else:
                df = pd.concat([df, ps], axis=0, ignore_index=True)

    return df

gl = get_game_log_of_player('203999')

gl

# print(gl)
# print(gl.shape)
        
# Nikola Joki
# stats = playergamelogs.PlayerGameLogs(player_id_nullable='203999', season_nullable="2015-16")
# stats = playergamelogs.PlayerGameLogs(player_id_nullable='203999', season_nullable="2021-22", season_type_nullable="Regular Season")
# pstats = playergamelogs.PlayerGameLogs(player_id_nullable='203999', season_nullable="2021-22", season_type_nullable="Playoffs")

# s = stats.get_data_frames()
# ps = pstats.get_data_frames()


# print(s[0])
# print(ps[0])
```
